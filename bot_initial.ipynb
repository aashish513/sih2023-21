```python
def load_json_file(filename):
    with open(filename) as f:
        file = json.load(f)
    return file

filename = '/content/intents.json'

data_file = load_json_file(filename)
```

```python
data = data_file['intents']
```

```python
i = 0
for intent in data_file['intents']:
  for pattern in intent['patterns']:
    row = [pattern,intent['tag']]
    raw_df.iloc[i] = row
    i+=1
```
